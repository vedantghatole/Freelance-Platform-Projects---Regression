Detecting Outliers

```python
# checking outliers
sns.boxplot(x=df['Budget'])
```

```python
# Calculate the upper and lower limits
Q1 = df['Budget'].quantile(0.25)
Q3 = df['Budget'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(df['Budget']>=upper)[0]
lower_array = np.where(df['Budget']<=lower)[0]
```

```python
# Removing the outliers
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
```

```python
def convert_to_usd(data):
    conversion = {'EUR': 1.07, 'GBP': 1.24, 'USD': 1}
    data['Budget'] = [data.loc[i,'Budget'] * conversion[data.loc[i,'Currency']] for i in data.index]
    data.drop(columns=['Currency'] , axis=1, inplace=True)
    return data
```

2. Use machine learning to create clusters of similar projects.


```python
df1 = pd.DataFrame(df)
```

```python
df3=df['Budget']
```

```python
df2
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
categorical_features = df2.columns.tolist()
for col in categorical_features:
    df2[col] = encoder.fit_transform(df2[col])
```

```python
df2
```

```python
df4 = pd.concat([df2, df["Budget"]], axis = 1)

df4
```

```python
correlation = df4.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')

plt.title('Correlation between different fearures')
```

```python
from sklearn.preprocessing import StandardScaler
df4_std = StandardScaler().fit_transform(df4)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components= 3)

pr_cp = pca.fit_transform(df4_std)

pr_cp

pc = pd.DataFrame(pr_cp, columns = ["pc1", "pc2", "pc3"])

pc
```

```python
from sklearn.cluster import KMeans
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df4_std)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df4_std)
pred = kmeans.predict(df4_std)
pred[:5]
```

```python
pd.Series(pred).value_counts()
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.metrics import adjusted_rand_score
ari = adjusted_rand_score(pred, kmeans.labels_)
print(ari)
```

```python
# convert the dataframe to a np.array
# Visualizing the clusters for k=4
df4_std = np.array(df4_std) #that all
plt.scatter(df4_std[pred==0,0],df4_std[pred==0,1],s=50, c='purple',label='Cluster1')
plt.scatter(df4_std[pred==1,0],df4_std[pred==1,1],s=50, c='blue',label='Cluster2')
plt.scatter(df4_std[pred==2,0],df4_std[pred==2,1],s=50, c='green',label='Cluster3')
plt.scatter(df4_std[pred==3,0],df4_std[pred==3,1],
            s=50, c='yellow',label='Cluster4')
```

3. Create a regression model to predict the budget

```python
df.head()
```

```python
x = df.drop("Budget", axis = 1)
y = df["Budget"]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
categorical_features = x.columns.tolist()
for col in categorical_features:
    x[col] = encoder.fit_transform(x[col])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y,
                        train_size=0.8, random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
ytest
```

```python
ytrain= ytrain.values.reshape(-1,1)
ytest= ytest.values.reshape(-1,1)

from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.fit_transform(xtest)
ytrain = sc_y.fit_transform(ytrain)
ytest = sc_y.fit_transform(ytest)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression().fit(xtrain,ytrain)
lr
```

```python
lr.coef_
```

```python
print(lr.intercept_)
```

```python
# Predictions from our Model

predictions = lr.predict(xtest)
predictions= predictions.reshape(-1,1)
predictions
```

```python
result = {"actual": list(ytest) ,
                     "predictions": list(predictions)}

result = pd.DataFrame(result)
result
```

```python
import math
rmse = math.sqrt(mse)
rmse
```

4. Create a classification model to predict the value of the Type column.

```python
df
```

```python
df_new=pd.DataFrame(df)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
categorical_features = df_new.columns.tolist()
for col in categorical_features:
    df_new[col] = encoder.fit_transform(df_new[col])
```

```python
# Divide the dataset into independent and dependent variables
X = df_new.drop(["Type"],axis=1)
Y = df_new['Type']

#Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,
                                               random_state=42, shuffle=True)

#Data was splitted as 80% train data and 20% test data.

Y_train = Y_train.values.reshape(-1,1)
Y_test = Y_test.values.reshape(-1,1)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Logistic Regression
reg = LogisticRegression(random_state = 42)
accuracies = cross_val_score(reg, X_train, Y_train, cv=5)
reg.fit(X_train,Y_train)
Y_pred = reg.predict(X_test)

#Obtain accuracy
print("Train Score:",np.mean(accuracies))
print("Test Score:",reg.score(X_test,Y_test))
```

```python
result = {"actual": list(Y_test) ,
                     "predictions": list(Y_pred)}

result = pd.DataFrame(result)
result
```

```python
# Support Vector Classifier
svc = SVC(random_state = 42)
accuracies = cross_val_score(svc, X_train, Y_train, cv=5)
svc.fit(X_train,Y_train)
Y_pred = svc.predict(X_test)

#Obtain accuracy
print("Train Score:",np.mean(accuracies))
print("Test Score:",svc.score(X_test,Y_test))
```
